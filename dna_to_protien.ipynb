```python
def complementary_strand(dna):
    complement_map=str.maketrans('ATCG','TAGC')
    complement=dna.translate(complement_map)
    return complement

dna='ATGTCCTTTTTATGC'
complement=complementary_strand(dna)
print('Original DNA: ',dna)
print('complement strand: ',complement)
```
